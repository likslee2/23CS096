```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import csv 

options = webdriver.ChromeOptions()
options.add_argument("--lang=en")
```

<H1>CPU</H1>

```python
# Create lists to store data
results = []

driver = webdriver.Chrome(options=options, service=Service(ChromeDriverManager().install()))
web = f'https://pcbuilder.net/product/processor/'
driver.get(web)

# Wait for search results to load
items = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, '//tr[contains(@class, "items")]')))

for item in items:
    view_more_button = item.find_element(By.XPATH, './/span[@class="view-more"]')
    view_more_button.click()
    driver.implicitly_wait(5)
    # Store data in a dictionary
    print(item.text)
    result = {
        'Title': item.find_element(By.XPATH, './/div[@class="table_title"]').text,
        'Price': item.find_element(By.XPATH, './/td[@class="price"]').text[1:],
        'Brand': item.find_element(By.XPATH, './/div[@class="detail__value f_brand"]').text,
        'Model': item.find_element(By.XPATH, './/div[@class="detail__value f_model"]').text,
        'Cores': item.find_element(By.XPATH, './/div[@class="detail__value f_cores"]').text,
        'Threads': item.find_element(By.XPATH, './/div[@class="detail__value f_threads"]').text,
        'Socket Type': item.find_element(By.XPATH, './/div[@class="detail__value f_socket_type"]').text,
        'Base Speed': item.find_element(By.XPATH, './/div[@class="detail__value f_maximum_speed"]').text,
        'Turbo Speed': item.find_elements(By.XPATH, './/div[@class="detail__value f_maximum_speed"]')[1].text,
        'Integrated Graphics': item.find_element(By.XPATH, './/div[@class="detail__value f_integrated_graphics"]').text,
        'Memory Type': item.find_element(By.XPATH, './/div[@class="detail__value f_memory_type"]').text,
        # 'Memory Type': item.find_element(By.CSS_SELECTOR, '.detail__value f_integrated_graphics').get_attribute("innerHTML"),
        'Link': item.find_element(By.XPATH, './/a[@class="btn btn-primary component-btn"]').get_attribute('href')
    }

    # Append the dictionary to the results list
    results.append(result)

driver.quit()

with open('CpuDataset.csv', 'w', newline='', encoding='UTF8') as f:
    field_names = ['Title', 'Price', 'Brand', 'Model', 'Cores', 'Threads', 'Socket Type', 'Base Speed', 'Turbo Speed', 'Integrated Graphics', 'Memory Type', 'Link']
    writer = csv.DictWriter(f, fieldnames = field_names)
    writer.writeheader() 
    writer.writerows(results) 
```
